```python
x=float(input("Enter your Age:"))
if(x>0 and x<1):
  print("baby")
elif(x<=1 and x<=3):
  print("toddler")
elif(x>=4 and x<=12):
  print("kid")
elif(x>=13 and x<=19):
  print("teenager")
elif(x>=20 and x<=60):
  print("adult")
else:
  print("elder")
```

```python
year=int(input("Enter a year:"))
if(year%4==0):
  if(year%100==0):
    if(year%400==0):
      print(f"{year} is a leap year")
    else:
      print(f"{year} is not a leap year")
  else:
    print(f"{year} is a leap year")
else:
  print(f"{year} is not a leap year")
```
